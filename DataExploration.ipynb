Comparing length vs width of the sepal and petal 

```julia
begin
    scatter(title="Sepal lenght  vs Sepal width", xlabel="length", ylabel="width",
        df.SepalLength, df.SepalWidth, color="blue", label="sepal")
    scatter!(df.PetalLength, df.PetalWidth, color="red", label="petal")
end
```
